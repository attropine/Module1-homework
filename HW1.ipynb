```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = (8, 5)

#Частина 1: Математичні функції
def plot_trigonometric_functions():
    "Graph sin(x), cos(x), tan(x)"
    try:
        x = np.linspace(0, 2 * np.pi, 400)

        plt.figure()
        plt.plot(x, np.sin(x), label="sin(x)", color="blue")
        plt.plot(x, np.cos(x), label="cos(x)", color="green")
        plt.plot(x, np.tan(x), label="tan(x)", color="red")

        plt.ylim(-5, 5)
        plt.xlabel("x")
        plt.ylabel("Function value")
        plt.title("Trigonometric functions")
        plt.grid(True)
        plt.legend()
        plt.savefig("trigonometric_functions.png")
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

#Частина 2: Секторна діаграма 
def plot_specialties_pie():
    "Sector diagram of Specialties"
    try:
        specialties = [
            "Інформатика",
            "Математика",
            "Фізика",
            "Хімія",
            "Біологія",
        ]
        students = [45, 32, 28, 15, 20]

        explode = [0.1 if x == max(students) else 0 for x in students]

        plt.figure()
        plt.pie(
            students,
            labels=specialties,
            autopct="%1.1f%%",
            explode=explode,
            startangle=140,
        )
        plt.title("Distribution of students by specialties")
        plt.savefig("specialties_pie.png")
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

#Частина 3: Комбінована візуалізація
def plot_combined_visualization():
    "Combined visualization 2x2"
    try:
        fig, axs = plt.subplots(2, 2, figsize=(14, 10))

        # (1,1) Функції
        x = np.linspace(0, 2 * np.pi, 400)
        axs[0, 0].plot(x, np.sin(x), label="sin(x)")
        axs[0, 0].plot(x, np.cos(x), label="cos(x)")
        axs[0, 0].plot(x, np.tan(x))
        axs[0, 0].set_ylim(-5, 5)
        axs[0, 0].set_title("Trigonometric functions")
        axs[0, 0].grid()
        axs[0, 0].legend()

        # (1,2) Секторна діаграма
        specialties = ["Інформатика", "Математика", "Фізика", "Хімія", "Біологія"]
        students = [45, 32, 28, 15, 20]
        explode = [0.1 if x == max(students) else 0 for x in students]
        axs[0, 1].pie(
            students,
            labels=specialties,
            autopct="%1.1f%%",
            explode=explode,
        )
        axs[0, 1].set_title("Спеціальності")

        # (2,1) Гистограмма
        days = ["Пн", "Вт", "Ср", "Чт", "Пт", "Сб", "Нд"]
        temperature = [2, 4, 6, 5, 3, 1, 0]
        axs[1, 0].bar(days, temperature)
        axs[1, 0].set_title("Температура протягом тижня")
        axs[1, 0].set_ylabel("°C")

        # (2,2) Населення
        years = list(range(2014, 2024))
        population = [45, 44.5, 44, 43.5, 43, 42.5, 42, 41.5, 41, 40.5]
        axs[1, 1].plot(years, population, marker="o")
        axs[1, 1].set_title("Населення України (млн)")
        axs[1, 1].set_xlabel("Рік")
        axs[1, 1].set_ylabel("Млн осіб")
        axs[1, 1].grid()

        plt.tight_layout()
        plt.savefig("combined_visualization.png")
        plt.show()
    except Exception as e:
        print(f"Error: {e}")

#Частина 4: Додаткове ускладнення
def interactive_sin(amplitude):
    "Interactive graph of sin(x)"
    x = np.linspace(0, 2 * np.pi, 400)
    y = amplitude * np.sin(x)

    plt.figure()
    plt.plot(x, y)
    plt.title("Інтерактивний sin(x)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.show()

interact(
    interactive_sin,
    amplitude=FloatSlider(min=0.5, max=3, step=0.1, value=1),
)

#Виклик функцій
plot_trigonometric_functions()
plot_specialties_pie()
plot_combined_visualization()
```
